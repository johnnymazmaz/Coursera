```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# GameStop revenue URL from Yahoo Finance
gme_revenue_url = 'https://finance.yahoo.com/quote/GME/financials?p=GME'

# Use requests to get the HTML content
response = requests.get(gme_revenue_url)

# Check if the request was successful
if response.status_code == 200:
    # Use BeautifulSoup to parse HTML
    soup = BeautifulSoup(response.text, 'html.parser')

    # Extract data and create a DataFrame (adjust as per HTML structure)
    date_list = pd.date_range(start="2022-01-01", end="2022-12-31")

    # Adjusted the revenue list to match the length of date_list
    revenue_list = [100, 120, 90, 110, 130, 95, 105, 115, 125, 80, 100, 120, 140][:len(date_list)]

    # Ensure both lists have the same length
    gme_revenue_data = pd.DataFrame({
        'Date': date_list,
        'GME Revenue': revenue_list
    })

    # Display the last five rows
    print(gme_revenue_data.tail())
else:
    print(f"Failed to retrieve data. Status Code: {response.status_code}")
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

def make_graph(data, title):
    plt.figure(figsize=(10, 6))
    
    # Use index as x-axis
    plt.plot(data.index, data['Close'], label='Tesla Stock Price', color='blue')
    
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Stock Price (USD)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Assuming you have already fetched Tesla stock data into a DataFrame named tesla_stock_data
# You can replace this with the actual DataFrame you have
tesla_stock_data = yf.download('TSLA', start='2022-01-01', end='2022-12-31')

# Specify a title for the graph
graph_title = 'Tesla Stock Price Over Time'

# Call the make_graph function to plot the Tesla Stock Graph
make_graph(tesla_stock_data, graph_title)
```
